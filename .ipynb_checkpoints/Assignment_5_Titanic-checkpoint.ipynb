1. Show the age distribution in the data using a histogram.

```python
# Since We have detected missing data in the age column, we can use the mean value of the feature to impute the missing data
```

```python
# Selecting the Age column from the dataset
age = df['Age']
```

```python
age
```

```python
age.isnull().sum() # Number of missing values in the Age column
```

```python
# Using mean imputation method to replace null values in the Age column
df['Age'].fillna((df['Age'].mean()), inplace = True)
```

```python
#Check if the Age column has null values
age.isnull().sum() # Number of missing values in the Age column
```

```python
age.tail() # Index 888 replaced with mean age = 29.699118
```

```python
# Histogram plot 
fig = px.histogram(age, x='Age', nbins=16, title='Age Distribution of People Aboard the Titanic')


#  show plot 
fig.show()
```

5. Graph the value counts of the number of passengers who survived and did not survive
based on gender using a stacked bar graph.

```python
import seaborn as sns
import plotly.express as px

# Calculate the count of survivors and non-survivors based on gender
survival_counts = df.groupby(['Sex', 'Survived'])['Survived'].count().reset_index(name='count')

# Create a stacked bar chart using Plotly Express
fig = px.bar(survival_counts, x='Sex', y='count', color='Survived', barmode='stack', title = 'Stacked Bar Chart of Counts of Passengers who Survived/not Survived Based on Gender')

# Show the chart
fig.show()
```

```python
# More females (233) survived than males (109) 
# The proportion of females who survived is higher than the proportion of males who survived. 
# This could be explained by the "Women and Children First" code of conduct
# is a code of conduct whereby the lives of women and children were to be saved first in a life-threatening situation, 
# typically abandoning ship, when survival resources such as lifeboats were limited. However, it has no basis in maritime law. 
# wikipedia https://en.wikipedia.org/wiki/Women_and_children_first
```

7. Plot a bubble plot of the ages to the fare paid by each passenger categorizing whether
they survived or not. The size of each bubble should be determined by the passenger class
and the name of each individual as the hover name.

```python
# Load the dataset
df = pd.read_csv('Titanic Data.csv')
```

```python
# Create a bubble plot using Plotly Express
fig = px.scatter(df, x="Fare", y="Age", size="Pclass", color="Survived", hover_name="Name",
                 log_x=False, size_max=10,title= "Bubble Plot of the Age of Passengers who Survived /did not Survive to the Fare Paid")  # Log of Fare = False

# Show the plot
fig.show()
```

```python
# Create a bubble plot using Plotly Express
fig = px.scatter(df, x="Fare", y="Age", size="Pclass", color="Survived", hover_name="Name",
                 log_x=True, size_max=10, title = "Bubble Plot of the Age of Passengers who Survived /did not Survive to the Logarithm of Fare Paid") # Log of Fare

# Show the plot
fig.show()
```
